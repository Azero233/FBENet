```python
def Dcm2jpgEX(file_path,save_path,pname):
    c=[]
    names=os.listdir(file_path)
    # .dcm
    for name in names:
        index=name.rfind('.') # rfind()
        name=name[:index]
        c.append(name)

    
    for files in c:
        pircutre_path=file_path+"//"+files+".dcm"
        mdir = save_path+"//"+pname
        out_put=mdir+"//"+files+".jpg"
        try:
            
            ds=pydicom.read_file(pircutre_path)
            img=ds.pixel_array  # 
            if not os.path.exists(mdir):
                os.makedirs(mdir)
                imageio.imsave(out_put,img)
        except:
            print("")

    print("done")
Dcm2jpgEX("D:/myfile/2022-2023/dcm/0/CAI JIN/delay","D:/myfile/2022-2023/dcm/zhende0","CAI JIN")

pnames=os.listdir("D:/myfile/2022-2023/dcm/0")
for pname in pnames:
    Dcm2jpgEX("D:/myfile/2022-2023/dcm/0/"+pname+"/delay","D:/myfile/2022-2023/dcm/zhende0",pname)
```

```python
def Dcm2jpgEX2(file_path,save_path,pname):
    c=[]
    try:
        names=os.listdir(file_path)
        # .dcm
        for name in names:
            index=name.rfind('.') # rfind()
            name=name[:index]
            c.append(name)


        for files in c:
            pircutre_path=file_path+"//"+files+".dcm"
            out_put=save_path+"//"+pname+"_"+files+".jpg"
            try:
                ds=pydicom.read_file(pircutre_path)
                img=ds.pixel_array  # 
                imageio.imsave(out_put,img)
            except:
                print("")
    except:
        print("lack")
    print("done")
    

pnames=os.listdir("E:/myfile/2022-2023/dcm/Onpatient/val/patient0")
for pname in pnames:
    Dcm2jpgEX2("E:/myfile/2022-2023/dcm/Onpatient/val/patient0/"+pname+"/delay","E:/myfile/2022-2023/dcm/Onpatient/val/0",pname)
```

```python
import os
import random
import shutil
import cv2


def jpgCLAHE(file_path,save_path):
    c=[]
    try:
        names=os.listdir(file_path)
        # .dcm
        for name in names:
            index=name.rfind('.') # rfind()
            name=name[:index]
            c.append(name)


        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        for files in c:
            pircutre_path=file_path+"//"+files+".jpg"
            out_put=save_path+"//"+files+".jpg"
            try:
                img = cv2.imread(pircutre_path,0)
                img = clahe.apply(img)
                cv2.imwrite(out_put,img)
            except:
                print("")
    except:
        print("lack")
    print("done")

jpgCLAHE(r"D:\myfile\2022-2023\dcm\classify\valid\0",r"D:\myfile\2022-2023\dcm\tupianzengqiang\valid\0")
```
